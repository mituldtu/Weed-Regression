```python
broadleafdata = pd.read_excel("Broadleaf regression finalised.xlsx")
```

```python
import numpy as np
from numpy.polynomial.polynomial import polyfit


x = broadleafdata["Broadleaf Density of Weed 2022"]
y = broadleafdata["Grain Yield 2022"]

x1 = broadleafdata["Broadleaf Density of Weed 2023"]
y1 = broadleafdata["Grain Yield 2023"]

b, m = polyfit(x, y, 1)

plt.figure(figsize=(4,4))
plt.plot(x, y, '.',label = 'noneee')
plt.plot(x, b + m * x, '-',label = 'noneee')


b, m = polyfit(x1, y1, 1)

plt.plot(x1, y1, '.', label = 'noneee')
plt.plot(x1, b + m * x1, '--',label = 'noneee')



plt.show()
```
